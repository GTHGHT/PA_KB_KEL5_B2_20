### Visualisasi

- WordCloud Positif Label

```python
plt.figure(figsize=(45,45))
positif_label = data['Tweet'][data['sentimen'] == 1.0]
positif_wordcloud = WordCloud(stopwords=STOPWORDS, max_words=150, max_font_size=50, background_color='white').generate(str(positif_label))

plt.subplot(1, 3, 2)
plt.title("Positif Label")
plt.imshow(positif_wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

- WordCloud Negatif Label

```python
plt.figure(figsize=(45,45))
negatif_label = data['Tweet'][data['sentimen'] == 2.0]
negatif_wordcloud = WordCloud(stopwords=STOPWORDS, max_words=150, max_font_size=50, background_color='white').generate(str(negatif_label))

plt.subplot(1, 3, 2)
plt.title("Negatif Label")
plt.imshow(negatif_wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

- Pie Chart Sentimen Label

```python
data_pie = data['sentimen'].value_counts()
plt.figure(figsize =(10,10))
plt.pie(data_pie, startangle=90, autopct='%.1f%%')
plt.title("Sentimen label")
plt.show()
```
